# operators in python 

## 6 types of operators in python
1. Arithmetic operator
2. Assignment operator
3. Comparision operator
4. Logical operator
5. Membership operator
6. Identity operator

```python
#arithmetic operators - +,-,*,/,//,%
a=10
b=20
c=20
d=10
print (a+b)
print (b-c)
print (b*d)
print (d/a)
print (d//a)
print (d%a)
```

```python
#assignment operators- +=,-=,*=,/=,//=,%=
a=100
a+=50
print(a)
```

```python
b=200
b-=400
print(b)
```

```python
c= 50
c*=100
print(c)
```

```python
#comparision operators- >,<,>=,<=,!=,==
a= 10
b=20
c= 20
d=10
print(a>b)
print(b==c)
print(b<c)
print(b!=c)
print(b>=a)
print(b<=a)
```
